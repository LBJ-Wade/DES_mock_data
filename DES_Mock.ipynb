# Creating DES mock data for MGCosmoMC
This is just an exploratory example used to produce DES mock datasets for MGCosmoMC.

Let's import the libraries for this work

```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pil
```

```python
# loading the data
xip = np.loadtxt('./data/DES_1YR_final_xip.dat')
xim = np.loadtxt('./data/DES_1YR_final_xim.dat')
gammat = np.loadtxt('./data/DES_1YR_final_gammat.dat')
wtheta = np.loadtxt('./data/DES_1YR_final_wtheta.dat')
```

```python
# loading the MG theoretical predictions
mg1 = np.loadtxt('./input/DES_theory_vec_linear_weyl_mg1.dat')
```

```python
mg2 = np.loadtxt('./DES_theory_vec_linear_weyl_mg2.dat')
```

```python
# load the covariance matrices
invcov1 = np.loadtxt('./DES_inv_cov_mg1.dat')
invcov2 = np.loadtxt('./DES_inv_cov_mg2.dat')
```

```python
cov1 = np.linalg.inv(invcov1)
cov2 = np.linalg.inv(invcov2)
```

```python
data = np.loadtxt('./DES_vec_data_mg1.dat')
```

```python
mock1 = np.random.multivariate_normal(mg1, cov1)
```

```python
mock2 = np.random.multivariate_normal(mg2, cov2)
```

Now just some visual inspection

```python
plt.plot(np.arange(mock1.shape[0]), mock1, linestyle = '-', marker = 'o', alpha=0.5, color = 'C0', label='MG1')
plt.plot(np.arange(mock1.shape[0]), mg1, linestyle = '-', linewidth = 2, alpha=0.5, color = 'C0', label='MG1 theory')
plt.plot(np.arange(mock2.shape[0]), mock2, linestyle = '-', marker = 'x', alpha=0.5,  color = 'C1', label='MG2')
plt.plot(np.arange(mock1.shape[0]), mg2, linestyle = '-', linewidth = 2, alpha=0.5, color = 'C1', label='MG2 theory')
plt.plot(np.arange(data.shape[0]), data, linestyle = '-', marker = 'v', alpha=0.5,  color = 'C2', label='real data')
plt.xlim(-1,14.5)
plt.ylim(-2e-6,1e-5)
plt.legend(loc='upper right')
plt.xlabel('# data point')
plt.ylabel(r'$\xi_{+}^{11}$')
pil.savefig('./img/MockTest1.pdf', bbox_inches='tight')
pil.savefig('./img/MockTest1.png', bbox_inches='tight')
plt.show()
```
